```python
'''
Categorical Variable - Categorical Variable are not number.
Types -
1. Oridnal Variable - Category have some sort of order.
Like - Grade of schools (A,B,C,Fail)
2.Nomial Variable - Category is not related to each Other
Like - Conuntry Name - India,Pakistan
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
use_cols = ['id', 'purpose', 'loan_status', 'home_ownership']

data = pd.read_csv(
    'loan.csv', usecols=use_cols).sample(
        10000, random_state=44)  # set a seed for reproducibility

data.head()
```

```python
data.purpose.unique()      #Nominal Categorical Variable
```

```python
fig = data.purpose.value_counts().plot.bar()    #Value_counts()- To count the Number of Categories
```

```python
#Conclusion - Most of the People took loan for house improvment and to pay there credit card bills.
```

```python
data.loan_status.unique()
```

```python
fig = data.loan_status.value_counts().plot.bar()
```

```python
#Here it's looks most of the people alredy paid the bills.
```

```python
data.home_ownership.unique()    #Nominal Variable
```

```python
data.home_ownership.value_counts().plot.bar()
```
